#### Variables
- Variables are fundamental elements in programing used to store data that can we referenced and manipulated in a program. In python, Variables are created when you assign a value to them, and they do not explict declaration to reserve memory space. The declaration happens automatically when you assign a value to variables.

```python
a = 100
# a = variable
#100 = value
```

```python
# Declaring and assigning variables

age = 32
height = 6.1
name = "Deepak"
is_student = True

# Printing the variables

print("age:", age)
print("height:", height)
print("name:", name)
print("is_student:", is_student)
```

#### Naming conventions
- Variables names should be descriptive
- Variabl must start with a letter or an '_' and contains letters, numbers and underscores.
- Variables names case sensitive

```python
# Valid variable names

first_name = "Leo"
last_name = "Messi"
```

```python
# Case Sensitivity

name = "Deepak"
Name = "Kumar"

name == Name  # This will be False because variable names are case-sensitive
```

```python
# Understanding variable types

# Python is dynamically typed, meaning you don't need to declare variable types explicitly. The interpreter infers the type based on the assigned value.

age = 23          # Integer
height = 5.9     # Float
name = "Deepak"   # String
is_student = False  # Boolean
```

```python
# Type checking 

print(type(age))          # <class 'int'>
print(type(height))       # <class 'float'>
print(type(name))         # <class 'str'>
print(type(is_student))   # <class 'bool'>
```

```python
# Typpe Conversion

age_str = str(age)          # Convert integer to string
print(age_str, type(age_str))  # '23' <class 'str'>
```

```python
age = '25'
print(type(int(age)))
```

#### Dynamic Typing
- Python allows the type of a variable to change as the program excute

```python
var = 10
print(var,type(var))

var = 3.14
print(var,type(var))
```

```python
# Input

age = input("Enter your age: ")  # Input is always taken as a string
print("You entered age:", age, "of type", type(age))

age = int(input("Enter your age: "))  # Convert input to integer
print("You entered age:", age, "of type", type(age))
```
