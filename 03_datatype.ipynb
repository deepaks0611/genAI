#### Data Types

- Data types are classifiacation of data which tell the compiler how the programmer intends to use the data
They determine the types of operation that can be performed on the data, the values that can take and the amount of memory needed to store the data

#### Importance of data types in programming

- Data types ensure that data is stored in an efficient way.
They help in performing correct operations on data.
Proper use of data types can prevent errors and bugs in the program.

```python
# Integer

age = 25
print(type(age))
```

```python
# Floating Point data type

height = 5.9
print(type(height))
```

```python
# String Data Type

name = "Deepak Kumar"
print(type(name))
```

```python
# Boolean Data Type

is_student = True
print(type(is_student))
```

```python
a = 10
b = 10

type(a == b)  # Comparison operator returns boolean value
```

```python
## Common errors

result = "Hello" + 5
print(result)
```

```python
result = "Hello" + str(5)
print(result)
```
